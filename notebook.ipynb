```python
stock = {
    "pains au chocolat": 500,
    "croissants": 400,
    "baguettes": 500,
    "beignets": 200
}
```

```python
def visualiser(stock): 
    for k, v in stock.items():
        print(f"Produit : {k} - Quantité : {v} ")

def fabriquer(prod, quantity):
    stock.update({prod: quantity})
```

```python
fabriquer("chouquettes", 200)
visualiser(stock)
```
